```python
cars['cyl'] = cars['cyl'].astype('category')
cars['am'] = cars['am'].astype('category')
cars.head()
```

```python
# Linear models - Least Squares
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
```

```python
X_R1 = cars['wt']
y_R1 = cars['mpg']
```

```python
X_R1.shape
```

```python
linreg = LinearRegression().fit(np.array(X_R1).reshape(-1,1),y_R1)
```

```python
print(linreg.intercept_)
print(linreg.coef_)
```

```python
y_pred = linreg.predict(np.array(X_R1).reshape(-1,1))
```

```python
print('R-squared: ', r2_score(y_R1, y_pred))
```
